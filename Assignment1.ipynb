```python
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb 

bn=gum.BayesNet('Diabetes')
pregnancies=bn.add(gum.LabelizedVariable('pregnancies','pregnant?',2))
glucose=bn.add(gum.LabelizedVariable('Glucose','Glucose high?',2))
skinThickness=bn.add(gum.LabelizedVariable('SkinThickness','SkinThickness thick?',2))
insulin=bn.add(gum.LabelizedVariable('Insulin','Insulin high?',2))
BMI=bn.add(gum.LabelizedVariable('BMI','BMI high?',2))
diabetesPedigreeFunction=bn.add(gum.LabelizedVariable('DiabetesPedigreeFunction','DiabetesPedigreeFunction yes?',2))
age=bn.add(gum.LabelizedVariable('Age','Age high?',2))
bloodPressure=bn.add(gum.LabelizedVariable('BloodPressure','BloodPressure high?',2))
outcome=bn.add(gum.LabelizedVariable('outcome','outcome yes?',2))

bn.addArc(pregnancies,outcome)
bn.addArc(glucose,outcome)
bn.addArc(skinThickness,outcome)
bn.addArc(insulin,outcome)
bn.addArc(BMI,outcome)
bn.addArc(diabetesPedigreeFunction,outcome)
bn.addArc(age,outcome)
bn.addArc(bloodPressure,outcome)

bn
```
