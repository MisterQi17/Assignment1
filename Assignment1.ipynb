```python
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb 

bn=gum.BayesNet('Diabetes')
glucose=bn.add(gum.LabelizedVariable('Glucose','Glucose high?',2))
insulin=bn.add(gum.LabelizedVariable('Insulin','Insulin high?',2))
BMI=bn.add(gum.LabelizedVariable('BMI','BMI high?',2))
diabetesPedigreeFunction=bn.add(gum.LabelizedVariable('DiabetesPedigreeFunction','DiabetesPedigreeFunction yes?',2))
age=bn.add(gum.LabelizedVariable('Age','Age high?',2))
bloodPressure=bn.add(gum.LabelizedVariable('BloodPressure','BloodPressure high?',2))
diabetes=bn.add(gum.LabelizedVariable('Diabetes','outcome yes?',2))
treatment=bn.add(gum.LabelizedVariable('Treatment','treatment yes?',2))
ultility=bn.add(gum.LabelizedVariable('Ultility','ultility?',2))



bn.addArc(insulin,glucose)
bn.addArc(bloodPressure,glucose)

bn.addArc(glucose,diabetes)
bn.addArc(BMI,diabetes)
bn.addArc(diabetesPedigreeFunction,diabetes)
bn.addArc(age,diabetes)

bn.addArc(diabetes,ultility)
bn.addArc(treatment,ultility)

bn
```

```python
bn.cpt(bloodPressure).fillWith([0.5,0.5])
bn.cpt("BloodPressure")

bn.cpt(insulin).fillWith([0.5,0.5])
bn.cpt("Insulin")
bn.cpt(diabetesPedigreeFunction).fillWith([0.5,0.5])
bn.cpt("DiabetesPedigreeFunction")
# bn.cpt("Glucose").names

bn.cpt("Glucose")[{'BloodPressure': 0, 'Insulin': 0}] = [1, 0]
bn.cpt("Glucose")[{'BloodPressure': 0, 'Insulin': 1}] = [0.1, 0.9]
bn.cpt("Glucose")[{'BloodPressure': 1, 'Insulin': 0}] = [0.1, 0.9]
bn.cpt("Glucose")[{'BloodPressure': 1, 'Insulin': 1}] = [0.01, 0.99]
bn.cpt("Glucose")

bn.cpt(age).fillWith([0.5,0.5])
bn.cpt("Age")
bn.cpt(BMI).fillWith([0.5,0.5])
bn.cpt("BMI")

# bn.cpt(diabetes).fillWith([0.5,0.5])

# bn.cpt("Diabetes")[{'DiabetesPedigreeFunction':0, 'Glucose': 0, 'Age': 0, 'SkinThickness': 0, 'BMI': 0}] = [1, 0]
# bn.cpt("Diabetes")[{'DiabetesPedigreeFunction':0, 'Glucose': 1, 'Age': 0, 'SkinThickness': 0, 'BMI': 0}] = [1, 0]
# bn.cpt("Diabetes")[{'DiabetesPedigreeFunction':0, 'Glucose': 0, 'Age': 0, 'SkinThickness': 0, 'BMI': 0}] = [1, 0]
# bn.cpt("Diabetes")[{'DiabetesPedigreeFunction':0, 'Glucose': 0, 'Age': 0, 'SkinThickness': 0, 'BMI': 0}] = [1, 0]
# bn.cpt("Diabetes")[{'DiabetesPedigreeFunction':0, 'Glucose': 0, 'Age': 0, 'SkinThickness': 0, 'BMI': 0}] = [1, 0]
# bn.cpt("Diabetes")[{'DiabetesPedigreeFunction':0, 'Glucose': 0, 'Age': 0, 'SkinThickness': 0, 'BMI': 0}] = [1, 0]
# bn.cpt("Glucose")

bn.cpt(treatment).fillWith([0.5,0.5])
bn.cpt("Treatment")
# bn.cpt(ultility).fillWith([0.5,0.5])
```
